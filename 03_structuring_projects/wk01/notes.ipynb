# Machine learning strategy

* orthogonalization

    * focus on one perspective at the time (vs early stopping that affects two things at the same time)
    
    * performance on the train set
        * bigger net
        * different optimization algorithm
    * performance on the dev set
        * regularization
        * more data in train set
    * performance on the test set
        * more data in the dev set
    * performance in the real world
        * more data in the dev set
        * cost function

* objectives

    * dev set & single evaluation metric (is the new approach working?)
        * this speeds-up iteration
    * satisfying vs optimizing metric
        * optimizing accuracy for every instance of a model that was fitted in <= 100ms
    * single representative dev-set & single metric (!)

    * experimental design & data splits
        * small dataset (<10^4) -> cv, train-test split with ratios of 70/30
        * medium/large datasets (>10^6) -> train-test split with ratios of 97/3

    * Is the evaluation metric aligned with the problem?
        * standard & weighted metrics
        * tailored metrics
        * target metric definition vs optimizing performance (loss?)

* human-level performance

    * performance of ml solution
        * human-level error (progress usually fast until hle achieved)
            * get labeled data
            * manual error analysis
            * better analysis of bias/variance

        * Bayes optimal error (best possible error in mapping X->Y)
    
    * avoidable bias
        * human error as a proxy to Bayes error
        * human error much lower than achieved on the train set -> room for improvement on train set (avoidable bias)
        * human error comparable to the train set -> focus on dev set (variance)
    
    * improving performance
        * beyond human error -> less clear directions
        * natural perception tasks -> human good, others -> not that good
        * analyze differences between

            * human error, train error (bias)
                * bigger model
                * train for longer
                * optimizer
                * architecture/hyperparam search

            * dev error (variance)
                * more data
                * data augmentation
                * regularization
                * architecture/hyperparam search
